```python
def analysis(fname):
    wave = thinkdsp.read_wave(fname)
    wave.plot()
    wave.make_audio()
    spectrum = wave.make_spectrum()
    spectrogram = wave.make_spectrogram(256)
    return wave, spectrum, spectrogram
```

```python
analysis('Michael.wav')[1].plot(high=800,color='red')
analysis('Troy.wav')[1].plot(high=800,color='orange')
# analysis('Aaron.wav')[1].plot(high=800)
```
